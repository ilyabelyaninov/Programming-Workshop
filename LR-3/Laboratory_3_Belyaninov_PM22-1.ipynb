**Упражнение 2. Создание спрайтов в парадигме ООП.**

Второй этап разработки игры аркады:\
	создать спрайт, используя class Player(pygame.sprite.Sprite)\
	сделать движение спрайта по стрелкам клавиатуры (вверх, вниз, вправо, влево) \
	сделать движение ФОНА при достижении спрайта границ слева и справа.

```python
import pygame as pg
import sys

pg.init()

WINDOW_WIDTH = 800
WINDOW_HEIGHT = 600
BACKGROUND_SPEED = 1


# создание класса Player
class Player(pg.sprite.Sprite):
    def __init__(self, filename, hero_x=100, hero_y=250, x_speed=0, y_speed=0):
        pg.sprite.Sprite.__init__(self)
        self.image = pg.image.load(filename)
        self.rect = self.image.get_rect()
        self.hero_x = hero_x
        self.hero_y = hero_y
        self.rect.x = hero_x 
        self.rect.y = hero_y
        self.x_speed = x_speed
        self.y_speed = y_speed

    def update(self):
        self.rect.x += self.x_speed

        # проверка, чтобы персонаж не выходил за границы экрана по X
        if self.rect.left < 0:
            self.rect.left = 0
        elif self.rect.right > WINDOW_WIDTH:
            self.rect.right = WINDOW_WIDTH

        self.rect.y += self.y_speed

        # проверка, чтобы персонаж не выходил за границы экрана по Y
        if self.rect.top < 0:
            self.rect.top = 0
        elif self.rect.bottom > WINDOW_HEIGHT:
            self.rect.bottom = WINDOW_HEIGHT

            
window = pg.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pg.display.set_caption("Аркадная игра")

hero = Player('valorant.png')

background_img = pg.image.load('fon.png')
# изменение размера фона, исходя из разрешения экрана
background_img = pg.transform.scale(background_img, (WINDOW_WIDTH, WINDOW_HEIGHT))

background_x = 0
background_speed = 0
player_hero_speed = 1 # скорость игрока

# главный игровой цикл
running = True
while running:
    for event in pg.event.get():
        if event.type == pg.QUIT:
            running = False
        elif event.type == pg.KEYDOWN:
            # обработка нажатия клавиш
            if event.key == pg.K_UP:
                hero.y_speed = -player_hero_speed
            elif event.key == pg.K_DOWN:
                hero.y_speed = player_hero_speed
            elif event.key == pg.K_LEFT:
                hero.x_speed = -player_hero_speed
                background_speed = hero.x_speed*0.5  # установка скорости фона
            elif event.key == pg.K_RIGHT:
                hero.x_speed = player_hero_speed
                background_speed = hero.x_speed*0.5
        elif event.type == pg.KEYUP:
            # обработка отпускания клавиш
            if event.key in (pg.K_LEFT, pg.K_RIGHT):
                hero.x_speed = 0
                background_speed = 0
            elif event.key in (pg.K_UP, pg.K_DOWN):
                hero.y_speed = 0

    # обновление позиции игрока
    hero.update()

    # проверка положения персонажа относительно границ фона
    if (hero.rect.left <= 0) or (hero.rect.right >= WINDOW_WIDTH):
        try:
            background_x -= BACKGROUND_SPEED * hero.x_speed / abs(hero.x_speed)
        except:
            pass
        if background_x >= WINDOW_WIDTH:
            background_x -= WINDOW_WIDTH
        elif background_x <= -WINDOW_WIDTH:
            background_x += WINDOW_WIDTH

    # отрисовка фона
    window.blit(background_img, (background_x, 0))
    if background_x > 0:
        window.blit(background_img, (background_x - WINDOW_WIDTH, 0))
    else:
        window.blit(background_img, (background_x + WINDOW_WIDTH, 0))

    # отрисовка героя
    window.blit(hero.image, hero.rect)

    pg.display.flip()

pg.quit()
sys.exit()
```

**Упражнение 3. Создание группы спрайтов. Задача коллизии объектов.
Конечный этап разработки игры аркады:**
- создать два дополнительных спрайта и объединить все спрайты программы в группу спрайтов.
- сделать движение дополнительных спрайтов
- создать объект стрела с привязкой к главному спрайту
- при нажатии на пробел стрела должна лететь
- при коллизии объектов (стрела и дополнительный спрайт) должны исчезать и стрела и дополнительный спрайт.

список всех персонажей игры и процесс создания группы спрайтов:\
all_sprites = pygame.sprite.Group()\

all_sprites.add(hero2)\

```python
import pygame as pg
import sys

pg.init()
clock = pg.time.Clock()

WINDOW_WIDTH = 800
WINDOW_HEIGHT = 600
BACKGROUND_SPEED = 5

# создание класса Player
class Player(pg.sprite.Sprite):
    def __init__(self, filename, hero_x=100, hero_y=250, x_speed=0, y_speed=0):
        pg.sprite.Sprite.__init__(self)
        self.image = pg.image.load(filename)
        self.rect = self.image.get_rect()
        self.hero_x = hero_x
        self.hero_y = hero_y
        self.rect.x = hero_x  # установка позиции героя по X
        self.rect.y = hero_y  # установка позиции героя по Y
        self.x_speed = x_speed 
        self.y_speed = y_speed 

    # метод для обновления позиции игрока
    def update(self):
        self.rect.x += self.x_speed

        # проверка, чтобы персонаж не выходил за границы экрана по X
        if self.rect.left < 0:
            self.rect.left = 0
        elif self.rect.right > WINDOW_WIDTH:
            self.rect.right = WINDOW_WIDTH

        self.rect.y += self.y_speed

        # проверка, чтобы персонаж не выходил за границы экрана по Y
        if self.rect.top < 0:
            self.rect.top = 0
        elif self.rect.bottom > WINDOW_HEIGHT:
            self.rect.bottom = WINDOW_HEIGHT

# класс со стрелой            
class Arrow(pg.sprite.Sprite):
    def __init__(self, arrow):
        pg.sprite.Sprite.__init__(self)
        self.image = pg.Surface((10, 5))
        self.image.fill((255, 0, 0))
        self.rect = self.image.get_rect()
        self.rect.center = arrow.rect.center  # установка позиции стрелы по центру героя
        self.speed = 20

    # метод для обновления позиции стрелы
    def update(self):
        self.rect.x += self.speed
        if self.rect.x > WINDOW_WIDTH:
            self.kill()  # удаление стрелы, когда она выходит за пределы окна

        # проверка попаданий стрелы
        for sprite in pg.sprite.spritecollide(self, all_sprites, False):
            if isinstance(sprite, AdditionalSprite):
                # если стрела попала в доп героя
                sprite.kill()
                self.kill()

# класс с дополнительными героями                
class AdditionalSprite(pg.sprite.Sprite):
    def __init__(self, filename, x, y, speed):
        pg.sprite.Sprite.__init__(self)
        self.image = pg.image.load(filename)
        self.rect = self.image.get_rect()
        self.rect.x = x
        self.rect.y = y
        self.speed = speed
    
    # изменение напралений движения
    def update(self):
        if self.rect.top <= 0 and self.speed < 0:
            self.speed = abs(self.speed)

        elif self.rect.bottom >= WINDOW_HEIGHT and self.speed > 0:
            self.speed = -abs(self.speed)
        self.rect.y += self.speed
        

window = pg.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pg.display.set_caption("Аркадная игра")

hero = Player('valorant.png')

background_img = pg.image.load('fon.png')
# изменение размера фона, исходя из разрешения экрана
background_img = pg.transform.scale(background_img, (WINDOW_WIDTH, WINDOW_HEIGHT))

# переменная для хранения текущего сдвига фона
background_x = 0
background_speed = 0 # скорость фона
player_hero_speed = 5 # скорость игрока

hero2 = AdditionalSprite('valorant_2.jpg', 600, 200, 10)
hero3 = AdditionalSprite('valorant_3.jpg', 400, 300, -10)

# создание группы спрайтов
all_sprites = pg.sprite.Group()
all_sprites.add(hero)
all_sprites.add(hero2)
all_sprites.add(hero3)

# главный игровой цикл
running = True
while running:
    for event in pg.event.get():
        if event.type == pg.QUIT:
            running = False
        elif event.type == pg.KEYDOWN:
            if event.key == pg.K_UP:
                hero.y_speed = -player_hero_speed
            elif event.key == pg.K_DOWN:
                hero.y_speed = player_hero_speed
            elif event.key == pg.K_LEFT:
                hero.x_speed = -player_hero_speed
                background_speed = hero.x_speed * 0.5  # установка скорости фона
            elif event.key == pg.K_RIGHT:
                hero.x_speed = player_hero_speed
                background_speed = hero.x_speed * 0.5  # установка скорости фона
            elif event.key == pg.K_SPACE:
                # создание стрелы и добавление ее в группу
                arrow = Arrow(hero)
                all_sprites.add(arrow)
        elif event.type == pg.KEYUP:
            # обработка отпускания клавиш
            if event.key in (pg.K_LEFT, pg.K_RIGHT):
                hero.x_speed = 0
                background_speed = 0  # сброс скорости фона
            elif event.key in (pg.K_UP, pg.K_DOWN):
                hero.y_speed = 0

    # обновление позиции игрока и спрайтов
    all_sprites.update()

    # проверка положения персонажа относительно границ фона
    if (hero.rect.left <= 0) or (hero.rect.right >= WINDOW_WIDTH):
        try:
            background_x -= BACKGROUND_SPEED * hero.x_speed / abs(hero.x_speed)
        except:
            pass
        if background_x >= WINDOW_WIDTH:
            background_x -= WINDOW_WIDTH
        elif background_x <= -WINDOW_WIDTH:
            background_x += WINDOW_WIDTH

    # отрисовка фона
    window.blit(background_img, (background_x, 0))
    if background_x > 0:
        window.blit(background_img, (background_x - WINDOW_WIDTH, 0))
    else:
        window.blit(background_img, (background_x + WINDOW_WIDTH, 0))

    all_sprites.draw(window)

    pg.display.flip()
    clock.tick(60)

pg.quit()
sys.exit()
```
