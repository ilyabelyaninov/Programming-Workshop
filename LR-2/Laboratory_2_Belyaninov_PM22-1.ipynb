На основе представленного кода, создайте домик, верхушка крыши которого будет находится по центру экрана.

```python
import pygame as pg
import random

pg.init()

screen = pg.display.set_mode((800, 600), pg.RESIZABLE)
screen.fill([255, 255, 255])
pg.display.set_caption('Белянинов Илья Витальевич')

home = [[400, 300],  # Координаты вершин домика
        [550, 350],
        [550, 500],
        [250, 500],
        [250, 350]]

running = True  # Флаг,определяющий работу основного цикла программы
clock = pg.time.Clock()  # Создание объекта Clock для управления временем

while running:
    pg.draw.rect(screen, 'blue', (305, 390, 50, 50))  # Рисование окна
    color = [random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)]  # Генерация случайного цвета
    pg.draw.lines(screen, color, True, home, 2)  # Рисование линий, образующих домик
    pg.display.flip()  # Обновление экрана

    screen.fill([255, 255, 255])  # Заполнение экрана белым цветом перед

    for event in pg.event.get():  # Обработка событий
        if event.type == pg.QUIT:  # Если произошло закрытие окна
            running = False  #Завершение работы программы

    clock.tick(1)  # Частота обновления экрана

pg.quit()  # Закрыть окно крестиком
```

Упражнение 2. Анимация в pygame.\
Напишите программу анимации:\
	осуществите движение по горизонтали квадрата, прямоугольника, круга и треугольника.\
	при соприкосновении этих фигур с границами окна они должны менять свое направление на противоположное, а также менять цвет на случайный.\
	при клике мышью по фигуре должно происходить изменение цвета фигуры.

```python
import pygame as pg
import random

pg.init()

SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600

screen = pg.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT), pg.RESIZABLE)
screen.fill('white')
pg.display.set_caption('Белянинов Илья Витальевич')

# Функция для генерации случайного цвета
def rand_color():
    return [random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)]

class Figure:
    def __init__(self, x, y, color):
        self.x = x
        self.y = y
        self.color = color
        self.speed_x = random.uniform(-10, 10)  # Случайная скорость по оси X

    def draw(self):
        pass  # Отрисовка фигуры

    def move(self):
        pass  # Перемещение фигуры

    def change_color(self):
        self.color = rand_color() # Изменение цвета фигуры


class Square(Figure):
    def __init__(self, x, y, size, color):
        super().__init__(x, y, color)
        self.size = [size]
    
    def draw(self):
        pg.draw.rect(screen, self.color, (self.x, self.y, self.size[0], self.size[0]))  # Отрисовка квадрата

    def move(self):
        if self.x < 0 or self.x + self.size[0] > SCREEN_WIDTH:
            self.speed_x *= -1
            self.color = rand_color()  # Изменение цвета при столкновении с границами окна
        self.x += self.speed_x  # Перемещение квадрата


class Rectangle(Figure):
    def __init__(self, x, y, size1, size2, color):
        super().__init__(x, y, color)
        self.size = [size1, size2]  # Размеры прямоугольника

    def draw(self):
        pg.draw.rect(screen, self.color, (self.x, self.y, self.size[0], self.size[-1]))  # Отрисовка прямоугольника

    def move(self):
        if self.x < 0 or self.x + self.size[0] > SCREEN_WIDTH:
            self.speed_x *= -1
            self.color = rand_color()  # Изменение цвета при столкновении с границами окна
        self.x += self.speed_x  # Перемещение прямоугольника


class Circle(Figure):
    def __init__(self, x, y, radius, color):
        super().__init__(x - radius, y - radius, color)
        self.radius = radius
        self.size = [2*radius]

    def draw(self):
        pg.draw.circle(screen, self.color, [self.x + self.radius, self.y + self.radius], self.radius)  # Отрисовка круга

    def move(self):
        if self.x < 0 or self.x + self.size[0] > SCREEN_WIDTH:
            self.speed_x *= -1
            self.color = rand_color()  # Изменение цвета при столкновении с границами окна
        self.x += self.speed_x  # Перемещение круга

        
# Класс для треугольника
class Triangle(Figure):
    def __init__(self, x, y, size, color):
        super().__init__(x, y, color)
        self.size = [size]  # Размер стороны треугольника

    def draw(self):
        pg.draw.polygon(screen, self.color, [(self.x + self.size[0] // 2, self.y),
                                              (self.x + self.size[0], self.y + self.size[0]),
                                              (self.x, self.y + self.size[0])])  # Отрисовка треугольника

    def move(self):
        if self.x < 0 or self.x + self.size[0] > SCREEN_WIDTH:
            self.speed_x *= -1
            self.color = rand_color()  # Изменение цвета при столкновении с границами окна
        self.x += self.speed_x  # Перемещение треугольника


# Создание списка фигур
figures = [Square(100, 100, 50, rand_color()),
           Rectangle(100, 250, 80, 50, rand_color()),
           Circle(100, 400, 35, rand_color()),
           Triangle(100, 500, 60, rand_color())]

running = True
clock = pg.time.Clock()

# Основной цикл программы
while running:
    for event in pg.event.get():
        if event.type == pg.QUIT:
            running = False  # Завершение работы программы при закрытии окна
        elif event.type == pg.MOUSEBUTTONDOWN and event.button == 1: # Проверка щелчка левой кнопкой мыши
            pos = pg.mouse.get_pos()
            for figure in figures:
                # Проверка попадания щелчка мыши в область фигуры
                if (figure.x <= pos[0] <= figure.x + figure.size[0]) and (figure.y <= pos[1] <= figure.y + figure.size[-1]):
                    figure.change_color() # Изменение цвета фигуры по щелчку мыши

    screen.fill('white') # Очистка экрана
    for figure in figures:
        figure.move() # Перемещение фигур
        figure.draw() # Отрисовка фигур

    pg.display.flip()  # Обновление экрана
    clock.tick(60) # Ограничение частоты кадров
    
pg.quit() # Закрыть окно крестиком
```
